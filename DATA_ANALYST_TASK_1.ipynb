```python
import pandas as pd

df = pd.read_csv('marketing_campaign.csv', sep='\t')
print("Original dataset shape:", df.shape)
df.head()
```

```python
# 1. Handle Missing Values
missing_before = df.isnull().sum()
df = df.dropna()
missing_after = df.isnull().sum()

# 2. Remove Duplicate Rows
duplicates_before = df.duplicated().sum()
df = df.drop_duplicates()
duplicates_after = df.duplicated().sum()

# 3. Standardize Text Columns (example: Marital_Status)
df['Marital_Status'] = df['Marital_Status'].str.strip().str.lower()
df['Education'] = df['Education'].str.strip().str.lower()

# 4. Convert Date Column to Proper Format
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], dayfirst=True)

# 5. Rename Columns: lowercase, replace spaces with underscores
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# 6. Fix Data Types (example: income to float, kidhome to int)
df['income'] = df['income'].astype(float)
df['kidhome'] = df['kidhome'].astype(int)
df['teenhome'] = df['teenhome'].astype(int)
```

```python
df.to_csv('cleaned_marketing_data.csv', index=False)

files.download('cleaned_marketing_data.csv')
```

```python
print("✅ Cleaning Summary:")

print("\n➡️ Missing Values (before):\n", missing_before[missing_before > 0])
print("➡️ Missing Values (after):\n", missing_after[missing_after > 0])

print("\n➡️ Duplicates Removed:", duplicates_before)
print("\n➡️ Final Dataset Shape:", df.shape)
```
